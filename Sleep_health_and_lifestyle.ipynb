# Sleep_health_and_lifestyle 데이터 준비

   
## 데이터 불러오기
- pandas를 이용해서 데이터 불러오기

```python
import pandas as pd

sleep_data = pd.read_csv("/content/Sleep_health_and_lifestyle_dataset_final.csv")
sleep_data.head()
```

- 데이터 칼럼 확인

```python
sleep_data.columns
```

---

## 데이터 전처리
   
- '**Sleep Disorder**' 칼럼에 있는 결측 데이터를 확인해보았다.
    - NaN인 경우가 많았는데, 다음 칼럼 주제에 해당되지 않아서 적혀있지 않은 것 같다.
    - 'healthy'로 변경해주었다.

```python
sleep_data[sleep_data["Sleep Disorder"].isna()]
```

```python
sleep_data.fillna({"Sleep Disorder":"healthy"}, inplace= True)

sleep_data[sleep_data["Sleep Disorder"].isna()]
```

---

## 데이터 가공
- 이 데이터들 중에서 이번 프로젝트에서는 수치형 데이터인 'Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps' 만 이용한다.
   - 수치형 데이터가 아닌 나머지 칼럼은 제거한다.

```python
sleep_data = sleep_data.drop(['Gender', 'Occupation', 'BMI Category', 'Blood Pressure', 'Sleep Disorder'], axis = 1)

sleep_data.columns
```
